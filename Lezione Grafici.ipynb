# Grafici

https://matplotlib.org/

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 6)
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y, 'y')
plt.show()
```

```python
plt.plot(x,y, '-r')
plt.show()
```

```python
plt.plot(x,y, '--r')
plt.show()
```

```python
plt.plot(x,y, '*r')
plt.show()
```

```python
plt.plot(x,y, '--r')
plt.show()
```

```python
plt.plot(x,y, 'r')
plt.plot(x, y, '.b')
plt.show()
```

```python
plt.title('Il mio grafico')
plt.plot(x,y)
plt.plot(x,y, '*r')
plt.show()
```

```python
plt.title('Il mio grafico')
plt.xlabel('tempo (s)')
plt.ylabel('spazio (m)')
plt.plot(x,y)
plt.plot(x,y, '*r')
plt.show()
```

```python
plt.title('Il mio grafico')
plt.xlabel('tempo (s)')
plt.ylabel('spazio (m)')
plt.plot(x,y)
plt.plot(x,y, '*r')
plt.grid()
plt.show()
```

```python
plt.title('Il mio grafico')
plt.xlabel('tempo (s)')
plt.ylabel('spazio (m)')
plt.yticks(y)
plt.plot(x,y)
plt.plot(x,y, '*r')
plt.grid()
plt.show()
```

```python
plt.title('Il mio grafico')
plt.xlabel('tempo (s)')
plt.ylabel('spazio (m)')
plt.yticks(y)
plt.plot(x,y, label='Spezzata')
plt.plot(x,y, '*r', label='Campioni')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.title('Il mio grafico')
plt.xlabel('tempo (s)')
plt.ylabel('spazio (m)')

plt.yticks(y)
plt.plot(x,y, label='Spezzata')
plt.plot(x,y, '*r', label='Campioni')
plt.grid()
plt.legend()

plt.show()
```

```python
t = np.linspace(-3, 3, 300)
```

```python
y = np.sin(t)
```

```python
plt.title('Il mio grafico')
plt.xlabel('tempo (s)')
plt.ylabel('spazio (m)')

plt.plot(t,y)
plt.grid()
plt.savefig('grafico.png', dpi=150)
plt.show()
```
